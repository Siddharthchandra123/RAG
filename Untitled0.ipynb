```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/text_file/python_intro.txt", encoding="utf-8")
document=loader.load()
print(f"📄 loaded {len(document)} document")
print(f"content preview {document[0].page_content[:100]}")
```

```python
from langchain_community.document_loaders import DirectoryLoader
loader=DirectoryLoader(
    "../data/text_file",
    glob="**/*.txt",
    loader_cls=TextLoader,
    show_progress=True,
)

documents=loader.load()
```

```python
print(f"📂 Loaded {len(documents)} documents")
```

```python
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter,
    CharacterTextSplitter,
    TokenTextSplitter,
)
text=documents[0].page_content
text
```

```python
char_split=CharacterTextSplitter(
    separator="\n",
    chunk_size=100,
    chunk_overlap=20,
    length_function=len,
)

char=char_split.split_text(text)
```

```python
print(f"Number of chunks: {len(char)}")
char
```

```python
print(text)
```

```python
rec=RecursiveCharacterTextSplitter(
    separators=["\n\n","\n"," ",""],
    chunk_size=100,
    chunk_overlap=20,
    length_function=len,
)

rec_text=rec.split_text(text)
print("Created chunks :", len(rec_text))
rec_text[0]
```
